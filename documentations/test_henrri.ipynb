# Test de l'intégration Henrri

```python
import os
from henrri_connect import SyncHenrriClient
from dotenv import load_dotenv

load_dotenv(".env.henrri")

key = os.getenv("HENRRI_API_KEY", "")
secret = os.getenv("HENRRI_API_SECRET", "")
print(key, secret)
```

```python
client = SyncHenrriClient(key, secret)
```

```python
from henrri_connect.models import CustomerRequest

request = CustomerRequest(
    min_id=1,
    search="",
    from_date='2023-01-01',
    to_date='2026-06-09'
)
customers = client.customers.list_customers(request=request, with_selected_fields=False)
print(customers)
print(client._access_token)
```
