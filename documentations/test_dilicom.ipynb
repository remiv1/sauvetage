```python
import sys
from pathlib import Path
# Recherche ascendante de la racine du projet contenant 'db_models'
p = Path.cwd()
project_root = None
for parent in [p] + list(p.parents):
    if (parent / 'db_models').is_dir():
        project_root = parent
        break
if project_root:
    if str(project_root) not in sys.path:
        sys.path.insert(0, str(project_root))
    print('Added project_root to sys.path:', project_root)
else:
    print("Project root with 'db_models' not found; sys.path unchanged")
print('Python executable:', sys.executable)
print('CWD:', Path.cwd())

from dotenv import load_dotenv
from os import getenv
from db_models.repositories.dilicom import DilicomRepository

load_dotenv('./.env.dilicom')
host = getenv('DILICOM_HOST')
port = getenv('DILICOM_PORT')
id = getenv('DILICOM_ID')
password = getenv('DILICOM_SECRET')
```

```python
# Création d'un fichier de test
test_file_path = Path('test_upload.txt')
test_file_path.write_text('Ceci est un fichier de test pour l\'upload vers Dilicom.')
```

```python
with DilicomRepository() as repo:
    repo.upload(str(test_file_path), './I/test_upload.txt')
    test = repo.list_files(complete=True)
    test_complete = repo.list_files('./I', complete=True)

print('Files list after upload (complete):')
pprint(test)
print('Files list in ./I (complete):')
pprint(test_complete)
```
