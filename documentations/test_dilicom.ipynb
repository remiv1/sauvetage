# Feuille de test pour l'environnement Dilicom

Le but de ce notebook est de tester le fonctionnement de la bibliothèque dilicom-parser et de nettoyer les données extraites pour les rendre compatibles avec le format attendu.

## Import et instanciation du parser

```python
from dilicom_parser.transport import Connector
from dilicom_parser.parser import DistributorParser
dc = Connector(timeout=5, env_path='.env.dilicom')
dp = DistributorParser()
```

## Test de la connexion et de la liste des fichiers

```python
dc.print_config()
```

```python
dc.connect()
```

```python
files_list = dc.list_files("./O")
from pprint import pprint
pprint(files_list)
```

```python
dc.close()
```

```python
from pprint import pprint
with Connector() as server:
    test = server.list_files('./ARC')
print('Files list in ./ARC (complete):')
pprint(test)
```

## Parseur de fichiers

```python
with Connector() as server:
    files_list = server.download_all(archive=True)
lines = 0
for f in files_list:
    parsed_file = dp.parse_file(f)
    print(len(parsed_file.lines))
    lines += len(parsed_file.lines)
print('Parsed lines:')
print(lines)
```
