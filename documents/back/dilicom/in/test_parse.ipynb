```python
from pathlib import Path
from onixlib import Notice

for book in Notice.parse_full(Path("489084922.xml"), version="3.0").products:
    print(book.isbn, book.descriptive.title)
    print(book.authors[0].full_name)
    print(book.editor)
    print(book.price.ht, book.price.ttc, book.price.currency, book.price.vat_rate)
    for ps in book.product_supply:
        print(ps.supply_details)
        for sd in ps.supply_details:
            print(sd.supplier_name)
            for pr in sd.prices:
                print(pr.amount, pr.currency, pr.price_type)
```
