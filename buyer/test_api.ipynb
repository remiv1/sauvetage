```python
from woocommerce import API
from os import getenv
from dotenv import load_dotenv
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

load_dotenv()

rkey = getenv("WPWC_KEY_READ")
rsecret = getenv("WPWC_SECRET_READ")
wkey = getenv("WPWC_KEY_WRITE")
wsecret = getenv("WPWC_SECRET_WRITE")

rwcapi = API(
    url="https://localhost:9443/",
    consumer_key=rkey,
    consumer_secret=rsecret,
    wp_api=True,
    version="wc/v3",
    verify_ssl=False,
)
wwcapi = API(
    url="https://localhost:9443/",
    consumer_key=wkey,
    consumer_secret=wsecret,
    wp_api=True,
    version="wc/v3",
    verify_ssl=False,
)
```

```python
from pprint import pprint

response = rwcapi.get("orders").json()

pprint(response)
```

```python
metadatas = response.get("meta_data", [])
customer_type = next((
            (meta.get("value") for meta in metadatas if meta.get("key") == "_billing_wooccm10")
        ), "").lower()
print(f"Customer type: {customer_type}")
```

```python
# Créer un nouveau produit
new_product = {
    "name": "Nouveau Produit Test",
    "type": "simple",
    "regular_price": "29.99",
    "description": "Ceci est un produit de test créé via API"
}

response_post = rwcapi.post("products", new_product)
print(f"Status POST: {response_post.status_code}")

if response_post.status_code in [200, 201]:
    product_data = response_post.json()
    print("\nProduit créé avec succès!")
    print(f"ID: {product_data['id']}")
    print(f"Nom: {product_data['name']}")
    print(f"Prix: {product_data['regular_price']}")
else:
    print(f"Erreur: {response_post.text}")
```
